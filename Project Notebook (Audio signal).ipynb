```python
class SoundDataset(Dataset):
    
    ''' Instantiate Audio dataset object. '''
    
    def __init__(self, root : str, length : int = 32000,
                 transforms = None, partition : str = 'train'):
        
        self.root = root
        self.partition = partition
        self.transforms = transforms
        
        if not self.transforms:
            self.transforms = T.Lambda(lambda x: x + torch.randn(tuple(x.shape))/10)
            
        self.length = length
        self.files = os.listdir(self.root)
        
        self.label_int = {'B' : 0,
                          'I' : 1,
                          'F' : 2
                         }
        
        self.int_label = {(v, k) for (k, v) in self.label_int.items()}
    
    def __len__(self):
        return len(self.files)
    
    def __getitem__(self, index : int):
        fname = self.files[index]
        f = os.path.join(self.root, fname)
        
        file, _ = torchaudio.load(f)
        
        chance = torch.rand(1).item()
        
        if chance >= 0.5 and self.partition == 'train':
            file_2 = self.transforms(file.roll(torch.randint(low = -20, high = 21, size = (1,)).item()))
        else:
            file_2 = torch.clone(file).roll(torch.randint(low = -20, high = 21, size = (1,)).item())
        
        file = torch.cat([file, file_2], dim = -1)
        file = file[:, :self.length]
        
        return file, self.label_int[fname[0]]
```

```python
def conv_block(in_c, out_c, kernel_size = 3, pool_size = 4, padding = 1, stride = 1):
    
    ''' 
    Blueprint for Convolution block.
        
        Format: 
            conv->conv->maxpool->bn->relu->dropout
    '''
    model = nn.Sequential(nn.Conv1d(in_c, out_c, kernel_size = kernel_size, stride = stride, padding = padding),
                          nn.Conv1d(out_c, out_c, kernel_size = kernel_size, stride = stride, padding = padding),
                          nn.MaxPool1d(pool_size),
                          nn.BatchNorm1d(out_c, momentum = 0.8),
                          nn.ReLU(),
                          nn.Dropout(0.4)
                         )
    return model
```

```python
class AudioModel(nn.Module):
    
    ''' Model architecture for audio classification. '''
    
    def __init__(self, in_c, kernel_size, pool_size, channels):
        super(AudioModel, self).__init__()
        self.in_channels = in_c
        self.channels = channels
        self.kernel_size = kernel_size
        self.pool_size = pool_size
        
        module_list = []
        
        def conv_output(n, k, p, s):
            
            ''' Calculate output size for convolution operation. '''
            
            return (n - k + (2*p) + s)/s
        
        for c in self.channels:
            layers = []
            
            ### Paths 1 & 2
            layers.append(conv_block(in_c, kernel_size = self.kernel_size,
                                     pool_size = self.pool_size, out_c = c))
            layers.append(conv_block(in_c, kernel_size = self.kernel_size,
                                     pool_size = self.pool_size, out_c = c*2))
            
            module_list.append(layers)
            
            ### new input channel size for both paths
            in_c = c + c*2
            
        ### Save layers
        self.module_list_1, self.module_list_2 = zip(*module_list)
        
        ### Register layer parameters
        self.module_list_1 = nn.ModuleList(self.module_list_1)
        self.module_list_2 = nn.ModuleList(self.module_list_2)
        
        ### Initial sample size
        feature_length = train_dataset.dataset.length
        
        for i in range(len(self.channels)):
            feature_length = conv_output(n = feature_length, k = self.kernel_size, p = 1, s = 1)
            feature_length = conv_output(n = feature_length, k = self.kernel_size, p = 1, s = 1)
            
            feature_length /= self.pool_size

            #print(feature_length)
            
        #self.fc = nn.Linear(.length - (len(self.channels) * self.kernel_size), 3)
        self.fc = nn.Sequential(nn.Linear(in_c*int(feature_length), 3),
                                #nn.BatchNorm1d(1024, momentum = 0.8),
                                #nn.ReLU(),
                                #nn.Dropout(),
                                #nn.Linear(1024, 3)
                               )
        
        
        #s_list = [nn.Sequential(*layer) for layer in module_list]
        #self.model = nn.Sequential(*s_list)
        #self.fc = nn.Sequential(nn.Linear(self.channels[-1], ))
        
    def forward(self, x):
        for layers in zip(self.module_list_1, self.module_list_2):
            y1 = layers[0](x)
            y2 = layers[1](x)
            #print(f'y1 shape: {y1.shape}')
            #print(f'y2 shape: {y2.shape}')
            
            x = torch.cat([y1, y2], dim = 1)
        
            #print(x.shape)
        #b, feature_size = x.size(0), torch.prod(torch.tensor(x.shape[1:]))
        #print(f'Batch size: {b}\nFeature size: {feature_size}')
        #x = x.view(b, train_dataset.length - (len(self.channels) * self.kernel_size))
        #x = x.view(b, feature_size)
        #print(x.shape)
        
        x = x.view(-1, self.fc[0].in_features)
        
        return torch.log_softmax(self.fc(x), dim = -1)
```

```python
criterion = nn.NLLLoss()
lr = 1e-4
#optimizer = optim.Adam(model.parameters(), lr = lr)
optimizer = optim.Adam([{'params' : model.module_list_1.parameters(), 'lr' : lr/2},
                        {'params' : model.module_list_2.parameters(), 'lr' : lr/2} ,
                        {'params' : model.fc.parameters(), 'lr' : lr}
                       ], amsgrad = True)
lr_optim = optim.lr_scheduler.ExponentialLR(optimizer = optimizer, verbose = True, gamma = 0.7)

epochs = 20
history = {e + 1 : dict() for e in range(epochs)}

for epoch in range(epochs):
    ### Train loop
    train_acc = []
    model.train()
    for X, y in train_dl:
        optimizer.zero_grad()
        
        outputs = model(X)
        loss = criterion(outputs, y) + 0.1*torch.mean(torch.tensor([torch.norm(p, 1) for p in model.parameters()]))
        
        loss.backward()
        optimizer.step()
        
        correct = torch.sum(outputs.max(dim = -1)[1] == y)
        accuracy = correct/len(y)
        #print(f'Accuracy: {100*accuracy}%')
        train_acc.append(accuracy)
    
    lr_optim.step()
        
    ### Test loop
    model.eval()
    test_accuracies = []
    for X_, y_ in test_dl:
        preds = model(X_)
        acc = torch.sum(preds.max(dim = -1)[1] == y_)/len(y_)
        test_accuracies.append(acc)
    
    history[epoch + 1]['Train'] = 100 * np.mean(train_acc).item()
    history[epoch + 1]['Test'] = 100 * np.mean(test_accuracies).item()
    
    print(f'Epoch {epoch + 1:02d}\n  Train accuracy: {100 * np.mean(train_acc).item():.4f} %')
    print(f'  Test accuracy: {100 * np.mean(test_accuracies).item():.4f} %')
```
